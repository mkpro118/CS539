<div style="font-size: 17.5px;">
$\displaystyle X = \sum_{i=1}^{r}{\sigma_{i}u_{i}v_{i}^{T}} + \sum_{i=r+1}^{N}{\sigma_{i}u_{i}v_{i}^{T}}$<br>
$V_{r} = \begin{bmatrix} v_{1} & v_{2} & \cdots & v_{r}\end{bmatrix}$<br>
<br>
$X \times V_{r} = \sum_{i=1}^{r}{\sigma_{i}u_{i}v_{i}^{T}} \times V_{r} + \sum_{i=r+1}^{N}{\sigma_{i}u_{i}v_{i}^{T}} \times V_{r}$<br>
$X \times V_{r} = \sum_{i=1}^{r}{\sigma_{i}u_{i}v_{i}^{T}v_{i}} + \sum_{i=r+1}^{N}{\sigma_{i}u_{i}v_{i}^{T}v_{i}}$<br>
<br>
But the values $v_{r+1} \cdots v_{N}$ don't exist, so we can broadcast them to be $O$<br>
$X \times V_{r} = \sum_{i=1}^{r}{\sigma_{i}u_{i}v_{i}^{T}v_{i}} + O$<br>
$X \times V_{r} = U_{r}\Sigma_{r}V_{r}^{T}V_{r}$<br>
<br>
Since $V_{r}$ is an orthogonal matrix, $V_{r}^{T} \times V_{r} = I_{r}$<br>
$X \times V_{r} = U_{r}\Sigma_{r}I_{r}$<br>
<br>
So, $X \times V_{r} = U_{r}\Sigma_{r}$, and,<br>
$X \times V_{r} \times V_{r}^{T}= U_{r}\Sigma_{r}V_{r}^{T}$<br>
$XV_{r}V_{r}^{T} = U_{r}\Sigma_{r}V_{r}^{T}$
</div>

```python
# Question 2

# Setup
%matplotlib inline
from sklearn.decomposition import PCA
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('dark_background')

filename = 'EX_PCAdata.csv'

matrix = np.genfromtxt(filename, delimiter=',')
features = matrix[:, :-1]
labels = matrix[:, -1]

pca = PCA(n_components=2)

transformed = pca.fit(features).transform(features)
t1 = transformed[labels == 4]
t2 = transformed[labels == 7]

fig, ax = plt.subplots(1, 1, figsize=(8,8))

ax.scatter(t1[:, 0], t1[:, 1], c='lightblue')
ax.scatter(t2[:, 0], t2[:, 1], marker='^', c='hotpink')

plt.show()
```
