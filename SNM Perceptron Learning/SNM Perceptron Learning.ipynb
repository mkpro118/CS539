## Part(b)
<div style="font-size:16px;">
$\text{Using the first inequality,}$<br>
$w_{0} + 0.5w_{1} + 0.1w_{2} \lt 0$<br>
$w_{0} \lt -\left(0.5w_{1} + 0.1w_{2}\right)$<br>
<br>
$\text{Using the second inequality,}$<br>
$w_{0} + 0.1w_{1} + 0.5w_{2} \gt 0$<br>
$w_{0} \gt -\left(0.1w_{1} + 0.5w_{2}\right)$<br>
<br>
$\text{Combining the above, we get,}$<br>
$-\left(0.1w_{1} + 0.5w_{2}\right) \lt w_{0} \lt -\left(0.5w_{1} + 0.1w_{2}\right)$<br>
$-\left(0.1w_{1} + 0.5w_{2}\right) \lt -\left(0.5w_{1} + 0.1w_{2}\right)$<br>
$\left(0.1w_{1} + 0.5w_{2}\right) \gt \left(0.5w_{1} + 0.1w_{2}\right)$<br>
$w_{2} \gt w_{1}$<br>

$\text{Let's say }w_{2} = 20\text{,} w_{1} = 10 \text{  } \left(w_{2} \gt w_{1}\right) \text{ then,}$<br>
$w_{0} \lt -\left(0.5 \cdot 10 + 0.1 \cdot 20\right)$<br>
$w_{0} \lt -\left(5 + 2\right)$<br>
$w_{0} \lt -7$<br>
$w_{0} = -8$
