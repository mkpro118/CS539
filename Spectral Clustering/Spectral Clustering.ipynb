```python
# Question 1

# Setup
import numpy as np

A = np.array([
    [1, 2, 3],
    [2, 2, 1],
    [3, 1, 4],
])

D = np.diag(np.sum(A, axis=0))

L = D - A

Lrw = np.linalg.inv(D) @ L

Ls = (_ := np.linalg.inv(D ** 0.5)) @ L @ _

print(f'Degree Matrix\n{D}')
print(f'Laplacian Matrix\n{L}')
print(f'Normalized Random Walk Laplacian Matrix\n{Lrw}')
print(f'Normalized Symmetric Laplacian Matrix\n{Ls}')
```

```python
# Question 2

from sklearn.datasets import load_iris

data = load_iris()

X = data.data[:, :2]

from sklearn.cluster import SpectralClustering

cluster1 = SpectralClustering(n_clusters=3).fit(X)
cluster2 = SpectralClustering(n_clusters=2).fit(X)

%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('dark_background')

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 8))
ax1.scatter(X[:, 0], X[:, 1], c=cluster1.labels_)
ax2.scatter(X[:, 0], X[:, 1], c=cluster2.labels_)
plt.show()
```
