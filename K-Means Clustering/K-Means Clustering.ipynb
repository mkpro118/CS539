```python
# Question 2

import numpy as np
from sklearn.datasets import load_iris

data = load_iris()

X, y = data.data[:, :2], data.target
print(f'{X.shape = }')
print(f'{y.shape = }')
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('dark_background')

from sklearn.cluster import KMeans

n_clusters1 = 3
n_clusters2 = 2

k_means1 = KMeans(n_clusters=n_clusters1).fit(X)
k_means2 = KMeans(n_clusters=n_clusters2).fit(X)

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(12, 8))

ax1.scatter(X[:, 0], X[:, 1], c=k_means1.labels_, cmap='cool')
ax1.scatter(k_means1.cluster_centers_[:, 0], k_means1.cluster_centers_[:, 1],
            c='w', marker='x', linewidth=3, s=169)
ax2.scatter(X[:, 0], X[:, 1], c=k_means2.labels_, cmap='cool')
ax2.scatter(k_means2.cluster_centers_[:, 0], k_means2.cluster_centers_[:, 1],
            c='w', marker='x', linewidth=3, s=169)

plt.show()
```
